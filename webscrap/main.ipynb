```python
import requests
import json

# alert.json 的完整 URL
alert_json_url = "https://www.adcc.gov.hk/data/alert.json"

# 发送 GET 请求获取 JSON 数据
response = requests.get(alert_json_url)

if response.status_code == 200:
    # 解析 JSON 数据
    data = response.json()
    
    # 提取文章列表 (注意：数据在 'Alerts' 键下)
    articles = []
    alerts_list = data.get('Alerts', [])  # 获取 Alerts 列表
    
    for item in alerts_list:  # item 是一个字典
        title_zhhk = item.get('title', {}).get('zhHk', '')
        date = item.get('date', '未知日期')
        article_id = item.get('id', '无ID')
        
        # 构建文章详情页 URL
        detail_url = f"https://www.adcc.gov.hk/zh-hk/alerts-detail/alerts-{article_id}.html"
        
        articles.append({
            'title': title_zhhk,
            'date': date,
            'id': article_id,
            'url': detail_url
        })
    
    print(f"成功获取到 {len(articles)} 篇文章\n")
    
    # 打印前几篇文章的信息
    for i, article in enumerate(articles[:5], 1):
        print(f"{i}. {article['title']} ({article['date']}) -> {article['url']}")
else:
    print(f"请求失败，状态码: {response.status_code}")
```

```python
import requests
import json
from bs4 import BeautifulSoup
import re

# --- 第一步：从 alert.json 获取所有文章信息 ---
print("正在获取文章列表...")
alert_json_url = "https://www.adcc.gov.hk/data/alert.json"

# 发送 GET 请求获取 JSON 数据
response = requests.get(alert_json_url)

adcc_articles = [] # 初始化 adcc_articles 列表

if response.status_code == 200:
    # 解析 JSON 数据
    data = response.json()
    
    # 提取文章列表 (注意：数据在 'Alerts' 键下)
    alerts_list = data.get('Alerts', [])  # 获取 Alerts 列表
    
    for item in alerts_list:  # item 是一个字典
        # 从 title 字典中获取繁体中文标题 (zhHk)
        title_dict = item.get('title', {})
        title_zhhk = title_dict.get('zhHk', '')  # 优先获取繁体标题
        
        # 如果没有 zhHk，则尝试获取简体标题 (zhCn) 作为备用
        if not title_zhhk:
            title_zhhk = title_dict.get('zhCn', '无标题')
        
        date = item.get('date', '未知日期')
        article_id = item.get('id', '无ID')
        
        # 构建文章详情页 URL
        detail_url = f"https://www.adcc.gov.hk/zh-hk/alerts-detail/alerts-{article_id}.html"
        
        # 创建文章字典
        article_info = {
            'title': title_zhhk,
            'date': date,
            'id': article_id,
            'url': detail_url,
            'content': '' # 初始化 content 为空字符串，稍后填充
        }
        
        adcc_articles.append(article_info) # 将文章信息添加到 adcc_articles 列表
    
    print(f"成功获取到 {len(adcc_articles)} 篇文章的元数据\n")
    
    # --- 第二步：循环遍历 adcc_articles，获取每篇文章的正文 ---
    print("正在抓取每篇文章的正文...")
    for i, article in enumerate(adcc_articles, 1):
        print(f"正在处理第 {i} 篇: {article['title'][:30]}...") # 打印标题前30个字符作为进度
        
        try:
            # 发送 GET 请求获取文章页面内容
            response_detail = requests.get(article['url'])
            
            if response_detail.status_code == 200:
                soup = BeautifulSoup(response_detail.text, 'html.parser')
                
                # 提取正文 (找到 class 为 "_content-view" 的 div)
                content_container = soup.find('div', class_='_content-view')
            
                if content_container:
                    # 使用 get_text() 提取容器内的所有文本内容
                    content = content_container.get_text(separator='\n', strip=True)
                else:
                    # 如果找不到 _content-view 容器，则回退到提取所有元素
                    content_elements = soup.find_all(['p', 'div', 'span', 'ul', 'ol', 'li', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
                    content = "\n".join([elem.get_text(strip=True) for elem in content_elements])
                
                # 将获取到的正文内容添加到 adcc_articles 中对应文章的字典里
                adcc_articles[i-1]['content'] = content # 使用索引 i-1 来更新列表中的字典
                
            else:
                print(f"  - 请求文章失败，状态码: {response_detail.status_code}")
                adcc_articles[i-1]['content'] = "无法获取正文内容"
                
        except Exception as e:
            print(f"  - 抓取文章时出错: {e}")
            adcc_articles[i-1]['content'] = "抓取过程中发生错误"
    
    print("\n--- 所有文章正文抓取完成 ---\n")
    
    # --- 第三步：可以打印或保存结果 ---
    # 例如，打印前 3 篇文章的标题和内容预览
    for i, article in enumerate(adcc_articles[:3], 1):
        print(f"--- 第 {i} 篇: {article['title']} ---")
        print(f"日期: {article['date']}")
        print(f"ID: {article['id']}")
        print(f"URL: {article['url']}")
        print(f"内容预览:\n{article['content'][:500]}...") # 只打印前 500 字符
        print("-" * 50)
        
else:
    print(f"请求 alert.json 失败，状态码: {response.status_code}")
    print(response.text)

# 现在，'adcc_articles' 列表包含了所有文章的标题、日期、ID、URL 和 content
# 你可以将 'adcc_articles' 列表保存到文件或数据库中，或者进行后续分析
print(f"\nadcc_articles 列表已创建，共包含 {len(adcc_articles)} 篇文章的数据。")
```

```python
from typing import List

def split_into_chunks(doc_file: str) -> List[str]:
    with open(doc_file, 'r', encoding='utf-8') as file:  # 添加 encoding='utf-8' 以确保正确读取中文
        content = file.read()

    chunks = content.split("---\n\n")
    
    return chunks

# # 读取markdown文件的内容并且按文章切片
# chunks = split_into_chunks("hk01_scam_articles.md")

hk01_chunks = split_into_chunks("hk01_scam_articles.md")

# 读取adcc的chunks 
adcc_chunks = split_into_chunks("adcc_scam_articles.md")

# 合并两个列表
chunks = hk01_chunks + adcc_chunks

for i, chunk in enumerate(chunks):
    print(f"[{i}] {chunk[:200]}...")  # 打印每个 chunk 的前 200 个字符作为预览
    print("-" * 50) 
```

```python
from sentence_transformers import SentenceTransformer

embedding_model = SentenceTransformer("shibing624/text2vec-base-chinese")

def embed_chunk(chunk: str) -> List[float]:
    embedding = embedding_model.encode(chunk, normalize_embeddings=True)
    return embedding.tolist()


embedding = embed_chunk("测试内容")
print(len(embedding))
print(embedding)
```

```python
# 将切片完的数据依次转化为向量
embeddings = [embed_chunk(chunk) for chunk in chunks]

print(len(embeddings))
print(embeddings[0])
```

```python
import chromadb

chromadb_client = chromadb.EphemeralClient()
chromadb_collection = chromadb_client.get_or_create_collection(name="default")

def save_embeddings(chunks: List[str], embeddings: List[List[float]]) -> None:
    for i, (chunk, embedding) in enumerate(zip(chunks, embeddings)):
        chromadb_collection.add(
            documents=[chunk],
            embeddings=[embedding],
            ids=[str(i)]
        )
# 将转化为向量的数据存入向量数据库
save_embeddings(chunks, embeddings)
```

```python
def retrieve(query: str, top_k: int) -> List[str]:
    query_embedding = embed_chunk(query)
    results = chromadb_collection.query(
        query_embeddings=[query_embedding],
        n_results=top_k
    )
    return results['documents'][0]
    
# 用户提问，由于当前所有的数据都是由HK01的诈骗专栏读取，所有的文章都是繁体中文，因此对英文的问题不能做出准确的回复，需改进
query = "大学生最容易遇到的骗案"
# 召回的数量可以根据数据的多少进行适当调整，但是防止数据冗余，此处默认5
retrieved_chunks = retrieve(query, 10)

for i, chunk in enumerate(retrieved_chunks):
    print(f"[{i}] {chunk}\n")
```

```python
from sentence_transformers import CrossEncoder

# 第一次召回的数据优先等级不清楚，于是重排，按照相关程度从高到低排列
def rerank(query: str, retrieved_chunks: List[str], top_k: int) -> List[str]:
    cross_encoder = CrossEncoder('cross-encoder/mmarco-mMiniLMv2-L12-H384-v1')
    pairs = [(query, chunk) for chunk in retrieved_chunks]
    scores = cross_encoder.predict(pairs)

    scored_chunks = list(zip(retrieved_chunks, scores))
    scored_chunks.sort(key=lambda x: x[1], reverse=True)

    return [chunk for chunk, _ in scored_chunks][:top_k]
# 重排筛选出最高相关的三个数据，数量可以根据召回的数量适当更改
reranked_chunks = rerank(query, retrieved_chunks, 3)

for i, chunk in enumerate(reranked_chunks):
    print(f"[{i}] {chunk}\n")
```
